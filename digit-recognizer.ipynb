```python
#Importinhg Libraries
import numpy as np 
import pandas as pd 
import tensorflow as tf 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Conv2D,BatchNormalization,LeakyReLU,MaxPooling2D,Flatten,Dense
```

```python
#Dataset reading with pandas 
data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
data.head()
```

```python
data_label = data['label']

data = data.drop('label', axis=1)
data=tf.convert_to_tensor(data)
data= data.numpy().reshape(42000,28,28,1)
```

```python
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
test=tf.convert_to_tensor(test)
test = test.numpy().reshape(28000,28,28,1)
```

```python
#Building model
model=Sequential()
model.add(Conv2D(64,(5,5),strides=(2,2),padding='same',input_shape=(28,28,1)))
model.add(MaxPooling2D(2,2))
model.add(BatchNormalization())
model.add(LeakyReLU())

model.add(Conv2D(128,(5,5),strides=(2,2),padding='same'))
model.add(MaxPooling2D(2,2))
model.add(BatchNormalization())
model.add(LeakyReLU())

model.add(Conv2D(256,(5,5),strides=(2,2),padding='same'))
#model.add(MaxPooling2D(2,2))
model.add(BatchNormalization())
model.add(LeakyReLU())

model.add(Conv2D(512,(5,5),strides=(2,2),padding='same'))
#model.add(MaxPooling2D(2,2))
model.add(BatchNormalization())
model.add(LeakyReLU())

model.add(Flatten())

model.add(Dense(256,activation='relu'))
model.add(Dense(10,activation='sigmoid'))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(data,data_label,batch_size=32,epochs=10)
```

```python
predictions = model.predict(test)
```

```python
predictions = [np.argmax(i) for i in predictions]
print(predictions)
```

```python
imageIds = [i for i in range(1,len(predictions)+1)]
df = pd.DataFrame()
df.insert(0, 'ImageId', imageIds)
df.insert(1, 'Label', predictions)
df.head()
```

```python
df.to_csv('predictions.csv',index=False)
```
